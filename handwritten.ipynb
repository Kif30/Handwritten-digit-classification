#***handwritten digit classification***

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("olafkrastovski/handwritten-digits-0-9")

print("Path to dataset files:", path)
```

```python
#list all directries
import os
print(os.listdir(path))
```

```python
#display all the images or 5 images in each folder
import matplotlib.pyplot as plt
from PIL import Image
plt.figure(figsize=(12,10))
i=1
for class_name in os.listdir(path): #looping all folders
  class_path = os.path.join(path, class_name)#add main path to each folder
  if os.path.isdir(class_path):#if folder exist
    images=os.listdir(class_path)[:5]#take 5 images in each folder
    for img in images:
      img_path=os.path.join(class_path, img)#add path to each image
      img=Image.open(img_path)#open image

      plt.subplot(10,5,i)
      plt.imshow(img)
      plt.title(class_name)
      plt.axis('off')
      i+=1
plt.tight_layout()
plt.show()
```
